## Python as an object-oriented programming language

A **class** is a blueprint or template for creating objects. 
It defines the *attributes* (properties) and *methods* (functions) that its objects will have. 
For example:

```python
class Mammal:
    def __init__(self, name, breed):
        self.name = name     # attribute
        self.breed = breed   # attribute

    def sound(self):          # method
        print(f"{self.name} makes a sound!")
```

Here, `Mammal` is a class that specifies what a "mammal" object should contain (name, breed), and what it can do (make a sound).

A *method* is a function defined inside a class that operates on objects of that class.
The first parameter of a method is always `self`, which represents the instance calling the method.

A concrete realisation of a class is an **object**.
For example:

```python
my_cat = Mammal('Max', 'Maine Coon')
print(my_cat.name)
my_cat.sound() 
```

Here, `my_cat` is a specific instance of a class `Mammal`. It's a Main Coon called Max.  

A class can inherit the attributes and methods of another class. This is useful when you need to work with subcategories. For example, cats are mammals, so they can do all the same things that other mammals do, and more:

```python
class Mammal:
    def __init__(self, name, breed):
        self.name = name     # attribute
        self.breed = breed   # attribute

    def sound(self):          # method
        print(f"{self.name} makes a sound!")

class Cat(Mammal):

    def meow(self):
        print(f"{self.name} says meow!")
```

```python
c = Cat('Whiskers', 'British Shorthair')

print(c.name)
print(c.breed)
c.meow()
```

We can also redefine `Mammal` to allow custom sounds:

```python
class Mammal:
    def __init__(self, name, breed, sound):
        self.name = name
        self.breed = breed
        self._sound = sound

    def sound(self):
        print(f"{self.name} says {self._sound}")
```

```python
dog = Mammal('Bluey', 'Australian Shepherd', 'woof')
cat = Mammal('Whiskers', 'British Shorthair', 'meow')

dog.sound() 
cat.sound()
```

Classes can be very useful in research. Suppose you are doing a study on how teachers adapt their teaching methods to different classroom contexts. You conduct interviews, collect observation notes, and analyse reflective journals. You want to organise, analyse, and retrieve your data efficiently. Each participant (teacher) has multiple attributes (e.g., name, years of experience, subjects taught, coded themes) and associated qualitative data (e.g., interview transcripts, observation notes). You can create a `Teacher` class to model each participant as an object with attributes and methods. This would make it easier to store, retrieve, and manipulate data programmatically instead of juggling each teacher's data separately.

```python
class Teacher:
    def __init__(self, name, experience, subjects):
        self.name = name
        self.experience = experience
        self.subjects = subjects
        self.interviews = [] # list of interviews
        self.observations = [] # list of notes
        self.codes = [] # list of codes

    def add_interview(self, transcript):
        self.interviews.append(transcript)

    def add_observation(self, note):
        self.observations.append(note)

    def add_code(self, code):
        self.codes.append(code)

    def summary(self):
        print(f"Teacher: {self.name}")
        print(f"Experience: {self.experience} years")
        print(f"Subjects: {', '.join(self.subjects)}")
        print(f"Number of Interviews: {len(self.interviews)}")
        print(f"Codes: {', '.join(self.codes)}")
```

```python
t1 = Teacher('Alice', 10, ['Math', 'Science'])
t1.add_interview('Interview 1 transcript text...')
t1.add_observation('Observed use of group work in class.')
t1.add_code('Collaborative learning')

t1.summary()
```

Note that I've used methods `.append()` and `.join()` above. 
The `.append()` method belongs to the class `list` and it allows us to append an item to the end of a list, like so:

```python
colours = ['blue', 'red', 'green', 'yellow', 'purple']
print(colours)
colours.append('orange')
print(colours)
```

`.join()` comes from the `str` class and concatenates the elements of an iterable (like a list) into a single string, with the string it is called on used as the separator:

```python
words = ['Math', 'Science']
sentence = ', '.join(words)   # Joins with a comma followed by a space
print(sentence)   
```

## Lists

Lists are created using square brackets, like so:

```python
colours = ['blue', 'red', 'green', 'yellow', 'purple']
print(colours)
```

Lists can contain any kind of data, including a mix of different data types. For example:

```python
mixed_list = [1, 'Bluey', 3.679, False]
```

How can we access the second item in this list? In R, we would do an equivalent of:

```python
print(mixed_list[2])
```

But here, we get the third element, not the second. This is because, in Python, **counting starts from 0, not from 1**. 

```python
print(mixed_list[1])
```

We can also extract a range of items from a list; this is called **slicing**:

```python
print(mixed_list[1:3])
```

Take a look at this output. Is this what you expected to happen?

In Python, **a range of numbers is specified in the inclusive:exclusive format**. This means that the number *before* the colon (`:`) is included in the range, but the number *after* the colon is not included in the range. So, when we use `print(mixed_list[1:3])`, we are asking Python to extract item 1 (i.e., the second item in the list) up to, but not including, item 3. To extract items 1 to 3, including item 3, we need to write:

```python
print(mixed_list[1:4])
```

Lists can also contain other lists:

```python
my_pets = [ ['cat', 'Whiskers'], ['dog', 'Bluey'] ]
print(my_pets[0])
print(my_pets[0][1])
```

Another important thing about lists is that you do not need to create a for-loop to generate lists. Instead, you can use a method called **list comprehension** â€” this is an example of a feature that is very [Pythonic](https://www.askpython.com/python/examples/pythonic-way-of-coding). 

For example, to create a list of the squares of the numbers from 1 to 10, we could use a for-loop:

```python
squares = []
for x in range(1, 11):
    squares.append(x**2)
print(squares)
```

However, list comprehension would be much shorter and also more elegant:

```python
squares = [x**2 for x in range(1, 11)]
print(squares)
```

You can even combine list comprehension with a conditional statement. For example, let's make a list of squares for odd numbers only:

```python
squares = [x**2 for x in range(1, 11) if x % 2 != 0]
print(squares)
```

Now let's combine what we've learned about objects and lists:

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

students = [Student('Alice', 90), Student('Bob', 85), Student('Charlie', 92)]

high_scorers = [s.name for s in students if s.grade > 88]
print(high_scorers)
```

We created a `Student` class with the attributes `name` and `grade`, then made a list containing three objects of this class. Using list comprehension, we then generated a new list with the names of those students whose grade is above 88. Neat!

## Dictionaries

A dictionary is a **collection of key-value pairs**. Unlike lists, where you access items by their numeric index, dictionaries let you retrieve values using their associated keys.

```python
my_pets = {'cat':'Whiskers', 'dog':'Bluey'}
cat_name = my_pets['cat']
print(f'My cat is called {cat_name}.')
```

This example also uses an f-string (formatted string). F-strings allow you to insert variables directly into strings using braces `{}`.

Updating a dictionary is very easy. For example, we can modify the cat's name in the dictionary `my_pets` like so:

```python
my_pets['cat'] = 'Max'
print(my_pets)
```

We can also add a new pet to the list:

```python
my_pets['rat'] = 'Nibbles'
print(my_pets)
```

Or, if you want to add several rats:

```python
del my_pets['rat']
my_pets['rats'] = ['Nibbles', 'Oats', 'Squeak']
print(my_pets)
```

We can do this because, in Python, both lists and dictionaries are **mutable**.

We can change `students` from the example above into a dictionary by using student names as keys and `Student` objects as values:

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

students = {
    'Alice': Student('Alice', 90),
    'Bob': Student('Bob', 85),
    'Charlie': Student('Charlie', 92)
}

high_scorers = [s.name for s in students.values() if s.grade > 88]
print(high_scorers)
```

Instead of creating a list with the names of students who achieved high scores, we can create a dictionary that includes both their names and scores:

```python
high_scorers = {name: s.grade for name, s in students.items() if s.grade > 88}
print(high_scorers)
```

Can you read this code easily? This example illustrates that although list and dictionary comprehensions are powerful, they can quickly become difficult to read, especially for less experienced programmers.

## Regular expressions

```python
import re # imports the module needed to work with RegEx in Python

text = 'The winning numbers are 123, 456, and 789.'
numbers = re.findall(r"\d+", text)
print(numbers)
```

Here, we are trying to find all numbers used in a string `text`. The expression `\d+` means 'one or more digits', and the method `.findall()` returns all matches as a list. 

Suppose you need to extract all email addresses from a text. We can achieve this by using regular expressions in a simple function:

```python
import re

def extract_emails(text):
    """Extracts all email addresses from a string."""
    pattern = r'[\w\.-]+@[\w\.-]+\.\w+'
    return re.findall(pattern, text)
```

How does this work? The pattern looks for all occurrences where a sequence of one or more letters, numbers, underscores, dots, or hyphens (`[\w\.-]+`) is followed by the literal `@` symbol. After the `@`, it matches another sequence of one or more letters, numbers, underscores, dots, or hyphens (`[\w\.-]+`), followed by a dot and the top-level domain (one or more word characters) (`\.\w+`). Let's check if that works:

```python
text = 'Contact us at support@example.com, sales@example.org, or admin@my-site.co.uk.'
emails = extract_emails(text)
print(emails)
```

Great!

We can modify the code to find all words that start with the 'collab' element so that we can find all instances where related words 'collaboration' and 'collaborative' are used:

```python
class Teacher:
    def __init__(self, name, experience, subjects):
        self.name = name
        self.experience = experience
        self.subjects = subjects
        self.interviews = []  # list of interviews
        self.observations = []  # list of notes
        self.codes = []  # list of codes

    def add_interview(self, transcript):
        self.interviews.append(transcript)

    def add_observation(self, note):
        self.observations.append(note)

    def add_code(self, code):
        self.codes.append(code)

    def summary(self):
        print(f"Teacher: {self.name}")
        print(f"Experience: {self.experience} years")
        print(f"Subjects: {', '.join(self.subjects)}")
        print(f"Number of Interviews: {len(self.interviews)}")
        print(f"Codes: {', '.join(self.codes)}")

        ## up until now the code remains unchanged

        ## add a method to search for all words that start with a given string:

    def search_word(self, word):
        pattern = re.compile(rf"\b{re.escape(word)}\w*", re.IGNORECASE)
        found_in_codes = [c for c in self.codes if pattern.search(c)]

        print(f"Searching for words starting with '{word}':")
        print(f"Found in codes: {found_in_codes}")
```

## Data frames

A data frame is a concept you will be familiar with from R; it is essentially a table or a spreadsheet. In Python, a package for representing data frames is called [pandas](https://pandas.pydata.org/).

```python
import pandas as pd

# let's create a simple data frame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 35, 28],
    'Score': [88, 92, 79, 95]
}

df = pd.DataFrame(data)
print(df)
```

We can access a column (or multiple columns) by name, like so:

```python
print(df['Name'])
```

```python
print(df[['Name', 'Score']])
```

... or by index, like so:

```python
print(df.iloc[:, 0])
```

The first argument within the square brackets in the code above refers to rows: by using `:`, we are specifying that we want to extract *all* rows. The second argument refers to columns, and `0` represents the first column. As is standard for Python, the end index when using `.iloc` is *exclusive*:

```python
print(df.iloc[:, 0:1])
```

It's important to not confuse `.iloc` with a very similar method called `.loc`. In contrast to `.iloc`, which extracts columns based on their *position*, `.loc` uses index *labels*. Moreover, with `.loc`, the end index is *inclusive* (very confusing, I know!):

```python
print(df.loc[:, 'Name':'Score'])
```

```python
print(df.loc[:, 0:2]) # this produces an error
```

However, when working with *rows* (rather than columns), these methods behave more similarly - the main difference being the end index behaviour. This is because row indices are often just numbers, so the index and the label are effectively the same:

```python
print(df.loc[1])  # Using label (index)
print(df.iloc[1]) # Using position
```

You'll need to use `.loc` (not `.iloc`) when filtering with conditions. This is because `.iloc` works with positions, while conditions create a boolean mask tied to row *labels*:

```python
print(df.loc[df['Age'] > 25, 'Name'])
```

```python
print(df.iloc[df['Age'] > 25, 'Name']) # this produces an error
```

Of course, you can filter based on a condition without `.loc`:

```python
high_scores = df[df['Score'] > 90]
print(high_scores)
```

```python
high_scores = df.loc[df['Score'] > 90] # with .loc
print(high_scores)
```

You can add a new column like so:

```python
df['Passed'] = df['Score'] > 80
print(df)
```

... and modify an existing column like so:

```python
df['Score'] = df['Score'] + 5
print(df)
```

Now let's calculate some basic descriptive statistics from our DataFrame. For example, we can find the average score for students who passed the test and for those who didn't:

```python
print(df.groupby('Passed')['Score'].mean())
```

Here, we are using the `.groupby` method to split the DataFrame into groups based on whether the students passed the test and then apply the .`mean()` operation to each group.

You can apply lots of simple operations in a similar manner (e.g., `.min()`, `.max()`, `.sum()`, `.median()`, `.count()`, `.std()`), but for more complex operations (e.g., mathematical functions like log or exp, random number generation, advanced linear algebra) you will need to import a package called [NumPy](https://numpy.org/) (and see [this tutorial](https://github.com/jwcarr/sciprog25/blob/main/unit4.ipynb) for a list of other useful packages).

```python
import numpy as np
df['LogScore'] = np.log(df['Score'])
print(df)
```

Finally, you can combine DataFrames if you need to add more data:

```python
extra_data = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Subject': ['Math', 'Science']
})

merged_df = pd.merge(df, extra_data, on = 'Name', how = 'left')
print(merged_df)
```

In this example, we already had subject information for Alice and Bob stored in a separate DataFrame, and we used the `.merge` method to add these extra details to our original DataFrame. We told Pandas to merge based on the `Name` column, and we set `how = left` to keep all the rows from the left DataFrame (the one we started with) and bring in matching data from the right DataFrame where possible. If there's no match - as with Charlie and Diana, who had no subject data - Pandas fills the missing values with `NaN`. Very convenient!

One great feature of Pandas DataFrames is that they can handle complex, non-tabular data while remaining easy to work with. For example, we can create a DataFrame where each teacher has received multiple feedback comments from the headteacher.

```python
data = {
    'Teacher': ['Alice Johnson', 'Bob Smith', 'Charlie Evans', 'Diana Clark'],
    'YearsExperience': [5, 12, 8, 3],
    'Subjects': [
        ['Math', 'Physics'],
        ['English', 'History'],
        ['Biology', 'Chemistry'],
        ['Art', 'Design']
    ],
    'HeadteacherFeedback': [
        [
            'Great lesson planning.',
            'Could engage more with quieter students.'
        ],
        [
            'Strong knowledge of subject.',
            'Needs to improve punctuality.'
        ],
        [
            'Excellent student rapport.',
            'Very well organized.'
        ],
        [
            'Creative approach to lessons.',
            'Should improve classroom management.'
        ]
    ]
}
df = pd.DataFrame(data)
print(df)
```

As you can see, we have different types of data in each column.

We can *explode* feedback into separate rows to make it easier to read:

```python
df_exploded = df.explode('HeadteacherFeedback')
print(df_exploded[['Teacher', 'HeadteacherFeedback']])
```

We can search for keywords in feedback:

```python
df['HasPunctualityFeedback'] = df['HeadteacherFeedback'].apply(
    lambda feedbacks: any('punctuality' in fb.lower() for fb in feedbacks))
print(df[['Teacher', 'HasPunctualityFeedback']])
```

This code creates a new column called `HasPunctualityFeedback` in the DataFrame `df`. For each teacher, it then checks all their feedback comments from the `HeadteacherFeedback` column to see if any of those comments mention the word *punctuality*: we use the `.apply` method to go through each list of comments one by one and, inside the *apply*, a *lambda* function (a quick, unnamed function) runs for each list of comments. The expression `any('punctuality' in fb.lower() for fb in feedbacks)` checks every comment `fb` in the list, converts it to lowercase (`fb.lower()`), and looks for the word *punctuality* in each. 

We can export this DataFrame like so:

```python
df.to_csv('teacher_feedback_data.csv', index = False)
```

And load it back like so:

```python
df = pd.read_csv('teacher_feedback_data.csv')
print(df)
```

## Some words on visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (8, 5))
sns.barplot(x = 'Teacher', y = 'YearsExperience', data = df)
plt.title('Years of experience per teacher')
plt.xticks(rotation = 45)
plt.ylabel('Years of experience')
plt.tight_layout()
plt.show()
```

As you know, bar plots are a really bad way to visualise data... A violin plot is a better choice because it shows the full shape of the data. However, our current toy DataFrame only has four data points, which isn't enough to make this meaningful. To fix that, let's simulate a larger dataset with 100 teachers, including their years of experience and the number of subjects each teaches, and plot that instead.

```python
np.random.seed(42) # so that we can reproduce the result

# simulate data for 100 teachers
df = pd.DataFrame({
    'TeacherID': range(1, 101),  # IDs from 1 to 100
    'YearsExperience': np.random.normal(loc = 10, scale = 5, size = 100),  # mean 10 years, sd 5
    'NumSubjects': np.random.randint(1, 6, size = 100)  # between 1 and 5 subjects
})

# ensure no negative years of experience
df['YearsExperience'] = df['YearsExperience'].clip(lower = 0)

print(df.head())
```

Now let's compute the mean and one standard error of the mean for the variable `YearsExperience`:

```python
mean_exp = df['YearsExperience'].mean()
se_exp = df['YearsExperience'].std(ddof = 1) / np.sqrt(len(df))
```

Here, I set `ddof = 1` to specify the degrees of freedom. This makes the denominator in the standard deviation formula `n - 1`, which means we're calculating the sample standard deviation rather than the population standard deviation. In this context, the distinction isn't important, so it's fine to ignore it if it's unfamiliar.

Now let's start plotting!

```python
plt.figure(figsize = (6, 5))
sns.violinplot(y = 'YearsExperience', data = df, inner = None, color = 'green', alpha = 0.5,
              bw_method = 0.4)

# add the mean as a black dot
plt.scatter(0, mean_exp, color = 'black', zorder = 3, label = 'Mean')

# add 1 SE of the mean
plt.errorbar(x = 0, y = mean_exp, yerr = se_exp, fmt = 'none', ecolor = 'black', elinewidth = 2, 
             capsize = 5, zorder = 2, label = '1 SE of the mean')

# add axes labels etc
plt.title('Distribution of years of experience across teachers')
plt.ylabel('Years of Experience')
plt.xticks([])
plt.legend()
plt.tight_layout()
plt.show()
```

Have you noticed something weird? We don't have negative values in the `YearsExperience` column, but the plot suggests that we do! Let's make sure that the plot does not extend beyond the lowest and highest actual data values:

```python
plt.figure(figsize = (6, 5))
sns.violinplot(y = 'YearsExperience', data = df, inner = None, color = 'green', alpha = 0.5, 
               bw_method = 0.4, cut = 0)  # restrict the violin to min and max data points

# add the mean as a black dot
plt.scatter(0, mean_exp, color = 'black', zorder = 3, label = 'Mean')

# add 1 SE of the mean
plt.errorbar(x = 0, y = mean_exp, yerr = se_exp, fmt = 'none', ecolor = 'black', elinewidth = 2, 
             capsize = 5, zorder = 2, label = '1 SE of the mean')

# add axes labels etc
plt.title('Distribution of years of experience across teachers')
plt.ylabel('Years of experience')
plt.xticks([])
plt.legend()
plt.tight_layout()
plt.show()
```

Nice! Now that we can see the distribution of the data, it's clear that most teachers have around 10 years of experience, and the average is about 9 years. To gain an even deeper understanding of the data, let's add quartiles to the visualisation: this will show us how the data are spread across different ranges.

```python
plt.figure(figsize = (6, 5))
sns.violinplot(y = 'YearsExperience', data = df, inner = None, color = 'green', alpha = 0.5, 
               bw_method = 0.4, cut = 0)  # restrict the violin to min and max data points

# box and whiskers
sns.boxplot(y = 'YearsExperience', data = df, 
            width = 0.1, # narrower so it sits inside the violin
            showcaps = True, 
            boxprops = {'facecolor':'none', 'edgecolor':'red'},  # transparent fill, red border
            whiskerprops = {'color':'red'},
            medianprops = {'color':'red'},
            flierprops = {'marker':'o', 'markersize':3, 'markerfacecolor':'red'})

# add the mean as a black dot
plt.scatter(0, mean_exp, color = 'black', zorder = 3, label = 'Mean')

# add 1 SE of the mean
plt.errorbar(x = 0, y = mean_exp, yerr = se_exp, fmt = 'none', ecolor = 'black', elinewidth = 2, 
             capsize = 5, zorder = 2, label = '1 SE of the mean')

# add axes labels etc
plt.title('Distribution of years of experience across teachers')
plt.ylabel('Years of experience')
plt.xticks([])
plt.legend()
plt.tight_layout()
plt.show()
```

Now we can see that one quarter of the teachers have less than 7.5 years of experience, and half of them fall between 7.5 and 12.5 years. This plot is both informative but not really visually appealing... Let's see if we can make it a bit prettier.

```python
plt.figure(figsize = (7, 5))

sns.violinplot(y = 'YearsExperience', data = df, inner = None, color = '#4c9f70',  # soft teal
               alpha = 0.6, bw_method = 0.4, cut = 0)

# refined boxplot with dark slate gray outlines, no fill
sns.boxplot(y = 'YearsExperience', data = df, width = 0.12,
    showcaps = True,
    boxprops = {'facecolor':'none', 'edgecolor':'#2f4f4f', 'linewidth':1.5},
    whiskerprops = {'color':'#2f4f4f', 'linewidth':1.5},
    medianprops = {'color':'#2f4f4f', 'linewidth':2},
    flierprops = {'marker':'o', 'markersize':4, 'markerfacecolor':'#2f4f4f', 'markeredgecolor':'none'})

# mean as a slightly larger black dot with white edge for contrast
plt.scatter(0, mean_exp, color = 'black', s = 80, edgecolor = 'white', 
            linewidth = 1.5, zorder = 4, label = 'Mean')

plt.errorbar(x = 0, y = mean_exp, yerr = se_exp, fmt = 'none', ecolor = 'black', elinewidth = 2, 
             capsize = 6, zorder = 3, label = '1 SE of the mean')

# titles and labels with nicer fonts
plt.title('Distribution of years of experience across teachers', fontsize = 14, weight = 'semibold')
plt.ylabel('Years of experience', fontsize = 12)
plt.xticks([])
plt.legend(frameon = False, fontsize = 11)
plt.tight_layout()
plt.show()
```

This one looks better, but if you're familiar with R and ggplot2, you will appreciate how much easier it is to create beautiful plots there: much less effort and much better results!
